# Objetivo

Este notebook tem como objetivo a exploração de dados públicos sobre museus brasileiros coletados no site: https://dados.cultura.gov.br/dataset/series-historicas-cadastro-nacional-de-museus, a fim de identificar informações relevantes para possíveis atualizações no aplicativo e suas funcionalidades.

# Importação das bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Configrando tamanhos de display de dataframes

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

# Carregando tabela

```python
dados = pd.read_excel(r"../bases/CNM - Base e Dicionário de Dados 2017.xlsx")
print("Quantidade de linhas: ", dados.shape[0])
print("Quantidade de colunas: ", dados.shape[1])
dados.head()
```

```python
dados.info()
```

# Seleção de colunas relevantes

## Tratando nome das colunas

```python
display(dados.dropna())
colunas, dados = dados.iloc[15].tolist(), dados.drop(index=[15]) # Identificando nome das colunas
```

```python
display(colunas)
```

```python
dados.columns = colunas # Atribundo nome certo às colunas
dados.head()
```

## Tratando dados nulos

```python
dados = dados.dropna(how='all', ignore_index=True) # Excluindo todas as linhas que são inteiramente nulas
dados.head()
```

```python
dados = dados.drop(index=[0, 1]).reset_index(drop=True) # Excluindo linhas vazias
dados.head()
```

```python
print("Quantidade de colunas:", dados.shape[1])
dados.isna().sum()
```

```python
# Excluindo colunas que possuem muitos dados nulos
for k, v in dados.isna().sum().items():
    if v > 3000:
        dados.drop(columns=[k], inplace=True)

print("Quantidade de colunas:", dados.shape[1])
dados.isna().sum()
```

## Colunas de identificação

```python
colunas_relevantes = dados[['location', 'Nome']] # Preservando colunas interessantes, mas são específicas
colunas_relevantes.head()
```

```python
# Excluindo colunas que possuem muitos dados únicos, que por muitas das vezes são dados de identificação
for i in dados:
    if len(dados[i].value_counts()) >= 3000:
        dados.drop(columns=[i], inplace=True)

# Analisando colunas que ficaram e quantos dados únicos possuem
print("Quantidade de colunas:", dados.shape[1])
for i in dados:
    print('\n=====', i, ('====='))
    print(len(dados[i].value_counts()))
```

## Analisando dados de cada coluna

```python
for i in dados:
    print('=====', i, ('====='))
    print(dados[i].value_counts())
```


## Selecionando apenas as mais relevantes ao nosso aplicativo

```python
dados = dados[['Descrição Curta', 'Município (visitação)', 'Estado (visitação)', 'Status do Museu (situação de funcionamento)', 'O ingresso ao museu é cobrado?', 'O museu promove visitas guiadas?', 'Em caso positivo, especifique', 'Esfera', 'Ano de abertura', 'Tipo de museu', 'Propriedade do acervo', 'Temática do museu', 'Indique os instrumentos de documentação de acervo utilizados pelo Museu']]
dados.head()
```

```python
dados = pd.concat([dados, colunas_relevantes], axis=1) # Juntando dados que ficaram com as colunas relevantes que foram excluidas
dados.head()
```

## Ajustando localização

```python
dados[['Localização X', 'Localização Y']] = dados['location'].str.split(',', expand=True) # Separando longitude e latitude dos dados
dados = dados.drop(columns=['location'])
dados.head()
```

## Padronizando dados

```python
print(dados['Status do Museu (situação de funcionamento)'].unique())
dados['Status do Museu (situação de funcionamento)'] = dados['Status do Museu (situação de funcionamento)'].map({'aberto': 'Aberto', 'fechado': 'Fechado', 'implantacao': 'Implantação', 'extinto': 'Extinto'})
print(dados['Status do Museu (situação de funcionamento)'].unique())
```

```python
print(dados['O ingresso ao museu é cobrado?'].unique())
dados['O ingresso ao museu é cobrado?'] = dados['O ingresso ao museu é cobrado?'].map({'não': 'Não', 'sim': 'Sim', 'contribuição voluntária': 'Contribuição voluntária', 'Não informado': 'Não informado'})
print(dados['O ingresso ao museu é cobrado?'].unique())
```

```python
print(dados['O museu promove visitas guiadas?'].unique())
dados['O museu promove visitas guiadas?'] = dados['O museu promove visitas guiadas?'].map({'não': 'Não', 'sim': 'Sim', 'Não informado': 'Não informado'})
print(dados['O museu promove visitas guiadas?'].unique())
```

# Exploração visual

## Localização dos museus

##### Grande concentração de museus no sudeste e sul do país, principalmente em grandes metrópoles como São Paulo e Rio de Janeiro

## Museus por ano de abertura

```python
# Histograma para ano de fundação
plt.hist(dados.drop(index=(dados.loc[dados['Ano de abertura'] == 'Não informado']).index)['Ano de abertura'].astype(int), bins=20)
plt.xlabel('Ano de Fundação')
plt.ylabel('Frequência')
plt.title('Distribuição do Ano de Fundação dos Museus')
plt.show()
```

##### Grande número de museus abertos entre o ano de 1950 e o ano 2000

## Situação de funcionamento

```python
label = dados.drop(index=(dados.loc[dados['Status do Museu (situação de funcionamento)'] == 'Não informado']).index)['Status do Museu (situação de funcionamento)'].value_counts().keys() # Tratando dados para este gráfico. Excluindo dados com "Não informado" e guardando os valores únicos da coluna
valores = dados.drop(index=(dados.loc[dados['Status do Museu (situação de funcionamento)'] == 'Não informado']).index)['Status do Museu (situação de funcionamento)'].value_counts() # Tratando dados para este gráfico. Excluindo dados com "Não informado" e guardando frequência dos dados únicos da coluna

fg, ax = plt.subplots(figsize=(8,8))
wedges, texts = ax.pie(valores, startangle=90)

tamanho = [f'{status} - {count / sum(valores) * 100:.1f}%' for status, count in zip(valores.index, valores)] # Calculando porcentagem. Quantidade de dados pelo total para cada valor único
plt.legend(wedges, tamanho, title="Situação de Funcionamento", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1)) # Adicionando legenda com as porcentagens
plt.title('Situação de Funcionamento dos Museus')

plt.show()
```

##### A implantação de um museu é o processo de criação e desenvolvimento de uma instituição museológica

## Tipos de museus e temática

```python
coluna1 = 'Temática do museu'
coluna2 = 'Tipo de museu'

# Desconsiderando 'Não informado' e 'Não informada' para armazenar os dados nulos de cada coluna e clacular a quantidade que aparecem na base
label1 = dados.drop(index=(dados.loc[dados[coluna1] == 'Não informada']).index)[coluna1].value_counts().keys()
valores1 = dados.drop(index=(dados.loc[dados[coluna1] == 'Não informada']).index)[coluna1].value_counts()

label2 = dados.drop(index=(dados.loc[dados[coluna2] == 'Não informado']).index)[coluna2].value_counts().keys()
valores2 = dados.drop(index=(dados.loc[dados[coluna2] == 'Não informado']).index)[coluna2].value_counts()

# Criando subplots
fg, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))

fg.figure = (200, 50)

tamanho1, texto1 = ax1.pie(valores1, startangle=90)
tamanho = [f'{status} - {count / sum(valores1) * 100:.1f}%' for status, count in zip(valores1.index, valores1)] # Calculando porcentagem. Quantidade de dados pelo total para cada valor único
ax1.legend(tamanho1, tamanho, title=coluna1, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1)) # Configurando legenda

tamanho2, texto2  = ax2.pie(valores2, startangle=90)
tamanho = [f'{status} - {count / sum(valores2) * 100:.1f}%' for status, count in zip(valores2.index, valores2)] # Calculando porcentagem. Quantidade de dados pelo total para cada valor único
ax2.legend(tamanho2, tamanho, title=coluna2, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1)) # Configurando legenda

plt.title('Museus', loc='center')
plt.show()
```

##### Temática do museu aborda os temas das obras representadas nos museus e, tipo do museu, o formato como este museu é apresentado

## Temas dos museus por estado

```python
import matplotlib.pyplot as plt

contagem_status_tipo = dados.drop(index=(dados.loc[dados['Temática do museu'] == 'Não informada']).index).groupby(['Estado (visitação)', 'Temática do museu']).size().unstack()

contagem_status_tipo.plot(kind='bar', stacked=True, figsize=(12, 7))

plt.title('Temas dos museus por estado')
plt.xlabel('Estado')
plt.ylabel('Número de Museus')
plt.legend(title='Temática do museu', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

##### No estado de São Paulo é visto que, após história, os temas mais frequentes são 'Arte, arquitetura e linguística' e 'Ciências exatas, da terra, biológicas e da saúde'

## Temas por tipo de museu

```python
import matplotlib.pyplot as plt
import textwrap

dados_visu = dados.drop(index=(dados.loc[dados['Temática do museu'] == 'Não informada']).index)
dados_visu = dados_visu.drop(index=(dados_visu.loc[dados_visu['Tipo de museu'] == 'Não informado']).index)

contagem_status_tipo = dados_visu.groupby(['Tipo de museu', 'Temática do museu']).size().unstack()

contagem_status_tipo = contagem_status_tipo.div(contagem_status_tipo.sum(axis=1), axis=0) * 100

ax = contagem_status_tipo.plot(kind='bar', stacked=True, figsize=(12, 7))

ax.set_title('Temas dos museus por tipo de museu')
ax.set_xlabel('Tipo de Museu')
ax.set_ylabel('Quantidade em prcentagem do total de cada tipo')
ax.legend(title='Temática do museu', bbox_to_anchor=(1.05, 1), loc='upper left')
labels = [textwrap.fill(label, width=11).replace("/", " ") for label in contagem_status_tipo.index] # Configurando quebra de texto
ax.set_xticklabels(labels, rotation=0, ha='center')

plt.tight_layout()
plt.show()
```

# Conclusão

A análise exploratória da base de dados pública sobre museus cadastrados em território nacional, permitiu identificar novas informações estratégicas para possíveis atualizações futuras.

Primeiramente, o mapeamento geográfico dos museus revelou uma significativa concentração nas regiões Sul e Sudeste do país, com destaque para o estado de São Paulo. Essa constatação valida a escolha inicial de localização do lançamento do aplicativo, visto que essas regiões concentram uma grande parcela das instituições culturais, o que poderá facilitar a captação de usuários e a visibilidade da plataforma.

Além disso, observou-se que muitos museus já oferecem guias próprios de visitação. Esse dado sugere que, para tornar a funcionalidade de guia do aplicativo mais atraente e diferenciada, é essencial implementar recursos adicionais que valorizem e complementem a experiência oferecida. Funcionalidades exclusivas, como realidade aumentada, interações multimídia ou rotas temáticas, podem representar diferenciais competitivos relevantes.

Por fim, é notável que uma ampla quantidade dos museus cadastrados são instituições tradicionais focadas no tema histórico, alinhando-se diretamente ao nosso foco como cliente na organização e catalogação de obras. Essa predominância indica que o aplicativo terá um público-alvo robusto e bem segmentado, com potencial para impulsionar a valorização e o acesso a acervos históricos por meio de uma ferramenta moderna e acessível.
